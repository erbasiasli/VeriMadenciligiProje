```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import ElasticNet, SGDRegressor
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import torch
import torch.nn as nn
import torch.optim as optim

warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
pd.set_option('display.float_format', lambda x: '%.2f' % x)

print("Kütüphaneler yüklendi, akış başlıyor.")
```

```python
# Veri yükleme
df = pd.read_csv("ekip_odevi_ham_veri_30k.csv")

# 1. Sütun seçimi ve temizlik
cols_to_keep = ['model', 'year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize', 'marka']
df = df[cols_to_keep]

# 2. String temizliği ve sayıya çevirme
df['price'] = df['price'].astype(str).str.replace(r'[£,]', '', regex=True)
df['mileage'] = df['mileage'].astype(str).str.replace(r'[£,]', '', regex=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')

# 3. Kirli veri ve eksik veri temizliği
df = df[~df['model'].astype(str).str.contains("unclean")]
df.dropna(inplace=True)

# 4. Aykırı değer (Outlier) temizliği
df = df[(df['price'] < 140000) & (df['price'] > 500)]
df = df[df['year'] > 1998]

# 5. Özellik Mühendisliği (Car Age & Mileage per Year)
df['car_age'] = 2024 - df['year']
df['mileage_per_year'] = df['mileage'] / df['car_age']
df['mileage_per_year'] = df['mileage_per_year'].replace([np.inf, -np.inf], 0).fillna(df['mileage'])

# 6. Hedef değişken log dönüşümü
df['log_price'] = np.log1p(df['price'])

print(f"Temizlik bitti. Kalan veri: {df.shape[0]} satır.")
display(df.head())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Marka bazlı fiyatlar
avg_price = df.groupby('marka')['price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price.index, y=avg_price.values, palette='magma', ax=ax[0])
ax[0].set_title('Markalara Göre Ortalama Fiyatlar')
ax[0].tick_params(axis='x', rotation=45)

# Korelasyon matrisi
numeric_cols = df.select_dtypes(include=[np.number]).drop(['price'], axis=1)
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap='coolwarm', ax=ax[1])
ax[1].set_title('Özellikler Arası Korelasyon (Log Price Dahil)')

plt.tight_layout()
plt.show()
```

```python
print("Advanced modeller için veri hazırlanıyor (Untitled 3 Modelleri).")

# Temizlenmiş veriden kopya alıyoruz
df_adv = df.copy()

# Referans dosyadaki Label Encoding (cat.codes) yöntemi
cat_cols = df_adv.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_adv[col] = df_adv[col].astype('category').cat.codes

# Hedef değişken senin dosyadaki gibi log_price üzerinden gidecek (daha stabil sonuç için)
X_adv = df_adv.drop(['price', 'log_price'], axis=1)
y_adv = df_adv['log_price']

X_train_adv, X_test_adv, y_train_adv, y_test_adv = train_test_split(X_adv, y_adv, test_size=0.2, random_state=42)

print(f"Advanced veri seti hazır: Eğitim {X_train_adv.shape}, Test {X_test_adv.shape}")
```

```python
print(" modeller (ElasticNet, SGD, AdaBoost, GBM) Hyperparameter Tuning ile eğitiliyor.")

advanced_models = {
    "ElasticNet": (ElasticNet(random_state=42), {
        'model__alpha': [0.01, 0.1, 1.0],
        'model__l1_ratio': [0.2, 0.5, 0.8]
    }),
    "SGD Regressor": (SGDRegressor(random_state=42), {
        'model__max_iter': [1000, 2000],
        'model__penalty': ['l2', 'elasticnet']
    }),
    "AdaBoost": (AdaBoostRegressor(random_state=42), {
        'model__n_estimators': [50, 100],
        'model__learning_rate': [0.01, 0.1]
    }),
    "Gradient Boosting": (GradientBoostingRegressor(random_state=42), {
        'model__n_estimators': [100, 200],
        'model__max_depth': [3, 5],
        'model__learning_rate': [0.05, 0.1]
    })
}

advanced_results = []

for name, (model, params) in advanced_models.items():
    pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')), 
        ('scaler', StandardScaler()), 
        ('model', model)
    ])
    
    # GridSearchCV 
    grid_search = GridSearchCV(pipe, params, cv=3, scoring='r2', n_jobs=-1)
    grid_search.fit(X_train_adv, y_train_adv)
    
    best_model = grid_search.best_estimator_
    y_pred_log = best_model.predict(X_test_adv)
    
    # Skorları log'dan geriye çevirerek hesaplıyoruz 
    y_test_real = np.expm1(y_test_adv)
    y_pred_real = np.expm1(y_pred_log)
    
    advanced_results.append({
        "Model": name,
        "Yöntem": "GridSearch Optimized",
        "R2 Skoru": r2_score(y_test_real, y_pred_real), 
        "MAE": mean_absolute_error(y_test_real, y_pred_real),
        "Optimum Parametreler": str(grid_search.best_params_)
    })

# Sonuçları göster
df_adv_results = pd.DataFrame(advanced_results)
display(df_adv_results.sort_values(by="R2 Skoru", ascending=False).style.background_gradient(cmap="Blues"))
```

```python
# ÖZELLİK MÜHENDİSLİĞİ VE TÜRETME
print("Özellik mühendisliği başlatılıyor.")

# Üst hücreden gelen df'i kullanıyoruz
df_eng = df.copy()

# 1. Özellik Türetme
df_eng['car_age'] = 2024 - df_eng['year']
df_eng['mileage_per_year'] = df_eng['mileage'] / (df_eng['car_age'] + 1)

# 2. Ekstra temizlik (Duplicated veriler)
df_eng = df_eng.drop_duplicates().reset_index(drop=True)

# 3. Hedef Değişken Dönüşümü
df_eng['log_price'] = np.log1p(df_eng['price'])

print(f"İşlem sonrası veri seti boyutu: {df_eng.shape}")
display(df_eng.head())
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class ShallowNN(nn.Module):
    def __init__(self, input_dim):
        super(ShallowNN, self).__init__()
        self.fc = nn.Sequential(nn.Linear(input_dim, 64), nn.ReLU(), nn.Linear(64, 1))
    def forward(self, x): return self.fc(x)

class DeepNN(nn.Module):
    def __init__(self, input_dim):
        super(DeepNN, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, 128), nn.BatchNorm1d(128), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(128, 64), nn.BatchNorm1d(64), nn.ReLU(),
            nn.Linear(64, 32), nn.ReLU(),
            nn.Linear(32, 1)
        )
    def forward(self, x): return self.layers(x)

def train_pytorch(model, X_tr, y_tr, X_te, y_te, epochs=100):
    X_tr_t = torch.FloatTensor(X_tr).to(device)
    y_tr_t = torch.FloatTensor(y_tr.values).view(-1, 1).to(device)
    X_te_t = torch.FloatTensor(X_te).to(device)
    
    optimizer = optim.Adam(model.parameters(), lr=0.005)
    criterion = nn.MSELoss()
    
    model.train()
    for _ in range(epochs):
        optimizer.zero_grad()
        loss = criterion(model(X_tr_t), y_tr_t)
        loss.backward()
        optimizer.step()
    
    model.eval()
    with torch.no_grad():
        preds = model(X_te_t).cpu().numpy()
    return preds

print("Yapay Sinir Ağları mimarisi yüklendi.")
```

```python
# 1. Temizlik
advanced_final_results = [res for res in advanced_final_results if "PyTorch" not in res['Model']]

# 2. X ve y Hazırlığı
from sklearn.preprocessing import StandardScaler

# X verilerini ölçeklendirmezsek model öğrenemez (PyTorch zorunluluğu)
scaler_x = StandardScaler()
X_train_scaled = scaler_x.fit_transform(X_train_adv).astype(np.float32)
X_test_scaled = scaler_x.transform(X_test_adv).astype(np.float32)

# Sizin yönteminiz: y üzerinde StandardScaler uygulaması
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train_adv.values.reshape(-1, 1)).astype(np.float32)
# y_test'i sadece dönüştürüyoruz (fit etmiyoruz)
y_test_scaled = scaler_y.transform(y_test_adv.values.reshape(-1, 1)).astype(np.float32)

# 3. Model Eğitimi (train_pytorch_v2 fonksiyonunu kullandığını varsayıyorum)
print("Sadece y-ölçeklendirme odaklı eğitim başlıyor...")

# Modelleri tekrar kur (temiz başlangıç için)
input_size = X_train_adv.shape[1]
shallow_model = ShallowNN(input_size).to(device)
deep_model = DeepNN(input_size).to(device)

# Eğit (y_train_scaled veriyoruz)
shallow_preds_s = train_pytorch_v2(shallow_model, X_train_scaled, y_train_scaled, X_test_scaled, epochs=300)
deep_preds_s = train_pytorch_v2(deep_model, X_train_scaled, y_train_scaled, X_test_scaled, epochs=300)

# 4. Geri Dönüşüm (Tersine Çevirme)
for name, p_s in [("Shallow NN", shallow_preds_s), ("Deep NN", deep_preds_s)]:
    # Ölçeklendirmeyi geri al (Standard -> LogPrice)
    p_log = scaler_y.inverse_transform(p_s.reshape(-1, 1))
    
    # Log dönüşümünü geri al (LogPrice -> Gerçek Pound)
    p_real = np.expm1(p_log).flatten()
    y_real = np.expm1(y_test_adv.values).flatten()
    
    # Skorları hesapla
    r2 = r2_score(y_real, p_real)
    mae = mean_absolute_error(y_real, p_real)
    
    advanced_final_results.append({
        "Model": name + " (y-Scaled)",
        "Yöntem": "Target Scaling Focus",
        "R2 Score": r2,
        "MAE": mae,
        "RMSE": np.sqrt(mean_squared_error(y_real, p_real)),
        "Özellik Sayısı": input_size,
        "Optimum Parametre": "y-Scaler Active"
    })
# 1. Listedeki R2 skoru 0'dan küçük olan tüm hatalı denemeleri temizle
advanced_final_results = [res for res in advanced_final_results if res['R2 Score'] > 0]

# 2. Temizlenmiş listeyle tabloyu yeniden oluştur
df_final_clean = pd.DataFrame(advanced_final_results)

# 3. Sonuçları göster (Sadece başarılı modeller kalacak)
print("\n--- FİNAL MODEL KARŞILAŞTIRMA TABLOSU (TEMİZLENMİŞ) ---")
display(df_final_clean.sort_values(by="R2 Score", ascending=False).style.background_gradient(cmap="Greens"))
```

```python
def train_pytorch(model, X_tr, y_tr, X_te, y_te=None, epochs=100):
    # Girişleri Tensor'a çevirirken Numpy array olduklarından emin oluyoruz
    # Eğer zaten numpy ise direkt çevirir, Pandas ise .values ile çevirir
    X_tr_t = torch.FloatTensor(np.array(X_tr)).to(device)
    y_tr_t = torch.FloatTensor(np.array(y_tr)).view(-1, 1).to(device)
    X_te_t = torch.FloatTensor(np.array(X_te)).to(device)
    
    optimizer = optim.Adam(model.parameters(), lr=0.005)
    criterion = nn.MSELoss()
    
    model.train()
    for _ in range(epochs):
        optimizer.zero_grad()
        loss = criterion(model(X_tr_t), y_tr_t)
        loss.backward()
        optimizer.step()
    
    model.eval()
    with torch.no_grad():
        preds = model(X_te_t).cpu().numpy()
    return preds
```

```python

# 1. Veri Hazırlığı ve Ölçeklendirme (X ve y için)
# datasets içindeki verilerin (X_train_adv gibi) hazır olduğunu varsayıyoruz
scaler_x = StandardScaler()
scaler_y = StandardScaler()

# X ve y'yi ölçeklendiriyoruz (PyTorch için hayati önemde)
X_train_scaled = scaler_x.fit_transform(X_train_adv).astype(np.float32)
X_test_scaled = scaler_x.transform(X_test_adv).astype(np.float32)

y_train_scaled = scaler_y.fit_transform(y_train_adv.values.reshape(-1, 1)).astype(np.float32)
y_test_scaled = scaler_y.transform(y_test_adv.values.reshape(-1, 1)).astype(np.float32)

# Modellerin deneneceği güncel veri seti sözlüğü
current_datasets = {
    "Advanced_Scaled": (X_train_scaled, X_test_scaled)
}

final_results = []

for data_name, (X_tr, X_te) in current_datasets.items():
    print(f"\n>>> İşleniyor: {data_name}...")
    
    # --- 1. Geleneksel Modeller ---
    # Not: Geleneksel modeller ölçeklenmiş y yerine normal log_price (y_train_adv) ile daha iyi çalışır
    en = ElasticNet(alpha=0.01, l1_ratio=0.5, random_state=42).fit(X_tr, y_train_adv)
    sgd = SGDRegressor(max_iter=2000, random_state=42).fit(X_tr, y_train_adv)
    ada = AdaBoostRegressor(n_estimators=100, learning_rate=0.05, random_state=42).fit(X_tr, y_train_adv)
    gbm = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42).fit(X_tr, y_train_adv)
    
    # --- 2. PyTorch Modelleri ---
    # Ölçeklenmiş y_train_scaled kullanıyoruz
    p_shallow_s = train_pytorch(ShallowNN(X_tr.shape[1]).to(device), X_tr, y_train_scaled, X_te, y_test_scaled, epochs=200)
    p_deep_s = train_pytorch(DeepNN(X_tr.shape[1]).to(device), X_tr, y_train_scaled, X_te, y_test_scaled, epochs=300)
    
    # --- 3. Tahminleri Geri Dönüştürme ve Kaydetme ---
    # PyTorch tahminlerini scaler_y ile geri çeviriyoruz
    p_shallow_log = scaler_y.inverse_transform(p_shallow_s.reshape(-1, 1))
    p_deep_log = scaler_y.inverse_transform(p_deep_s.reshape(-1, 1))
    
    model_names = ["ElasticNet", "SGD", "AdaBoost", "GBM", "PyTorch Shallow", "PyTorch Deep"]
    predictions = [en.predict(X_te), sgd.predict(X_te), ada.predict(X_te), gbm.predict(X_te), p_shallow_log, p_deep_log]
    
    for name, pred in zip(model_names, predictions):
        y_real = np.expm1(y_test_adv.values).flatten()
        p_real = np.expm1(pred).flatten()
        
        # Olası NaN değerleri 0 ile doldur (Hata koruması)
        p_real = np.nan_to_num(p_real)
        
        final_results.append({
            "Veri Tipi": data_name,
            "Model": name,
            "R2 Score": r2_score(y_real, p_real),
            "MAE": mean_absolute_error(y_real, p_real)
        })

# 4. Sonuçları Göster
df_final = pd.DataFrame(final_results)
print("\nEğitim ve Karşılaştırma Tamamlandı.")
display(df_final.sort_values(by="R2 Score", ascending=False).style.background_gradient(cmap="RdYlGn"))
```
