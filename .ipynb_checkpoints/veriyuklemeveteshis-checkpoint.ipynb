```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt 
import seaborn as sns           

print("Kütüphaneler başarıyla yüklendi")
```

```python
print(" Veri Yükleme ve Teşhis (EDA) Başlatıldı ")
data = pd.read_csv("ekip_odevi_ham_veri_30k.csv")
print(f"\nVeri Boyutu: {data.shape}")
```

```python
print("\n--- Eksik Veri Sayıları (isnull) ---")
print(data.isnull().sum().sort_values(ascending=False))
# her sütundaki toplam boş (NaN) veri sayısını hesapladık en çok boş olandan aza göre sıraladık.
# 'mpg', 'tax' ve 'engineSize' gibi sütunlarda eksik veri olduğunu bize gösterdi.                                 
```

```python
original_rows = data.shape[0]
data.drop_duplicates(inplace=True)
print(f" Tekilleştirme yapıldı. {original_rows - data.shape[0]} adet kopya satır kaldırıldı.")
# Bu işlem, veri setimizin "tarafsız" olmasını sağlar ve modelimizin yanlış yönlendirilmesini (bias) engeller.
```

```python
main_cols = ['model', 'year', 'price', 'transmission', 'mileage', 
             'fuelType', 'tax', 'mpg', 'engineSize', 'marka']
data_cleaned = data[main_cols].copy()
# Veri setimizin "genişliğini" (boyutunu) 17 gürültülü sütundan,
# 10 adet temiz ve anlamlı öznitelik sütununa indirdik.
```

```python
data_cleaned['price'] = data_cleaned['price'].astype(str).str.replace(r'[ £",]', '', regex=True)
data_cleaned['price'] = pd.to_numeric(data_cleaned['price'], errors='coerce')
data_cleaned['mileage'] = data_cleaned['mileage'].astype(str).str.replace(r'[",]', '', regex=True)
data_cleaned['mileage'] = pd.to_numeric(data_cleaned['mileage'], errors='coerce')
print("Veride 'price' ve 'mileage' sütunları temizlendi ve sayısala dönüştürüldü.")
```

```python
initial_row_count = data_cleaned.shape[0]
# Mantıksız veya aşırı aykırı değerleri filtreledik.
data_cleaned = data_cleaned[(data_cleaned['price'] > 100) & (data_cleaned['price'] < 250000)]
data_cleaned = data_cleaned[data_cleaned['mileage'] > 10]
data_cleaned = data_cleaned[data_cleaned['year'] > 1995]
print(f"Anomali filtrelemesi ile {initial_row_count - data_cleaned.shape[0]} adet mantıksız satır kaldırıldı.")
```

```python
#'engineSize' 0 olanları 'eksik' (NaN) olarak işaretledik.
data_cleaned['engineSize'] = data_cleaned['engineSize'].replace(0, np.nan)
print("Veride'engineSize' = 0 olanlar 'NaN' (eksik veri) olarak işaretlendi.")
```

```python
print(f"[BİLGİ] Bölüm 1 tamamlandı. Veri boyutu: {data_cleaned.shape}")
```
